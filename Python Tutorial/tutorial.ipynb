```python
from typing import List

class Solution:
    def dfs(self, grid: List[List[int]], r: int, c: int) -> int:
        if r < 0 or r >= len(grid) or c < 0 or c >= len(grid[0]) or grid[r][c] == 0:
            return 0
        grid[r][c] = 0
        return (1 + self.dfs(grid, r + 1, c) + self.dfs(grid, r - 1,c) + self.dfs(grid, r, c + 1) + self.dfs(grid, r, c -1) )

    def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
        max_area = 0 
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                max_area = max(max_area, self.dfs(grid, i, j))
        return max_area
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def where_is(point):
    match point:
        case Point(x=0, y=0):
            print("Origin")
        case Point(x=0, y=y):
            print(f"Y = {y}")
        case Point(x=x, y=0):
            print(f"X = {x}")
        case Point():
            print("Somewhere else")
        case _:
            print("Not a point")

my_point = Point(1,1)
where_is(my_point)
```

```python
def ask_ok(prompt, retries=4, reminder='Please try again!'):
    while True:
        reply = input(prompt)
        if reply in {'y', 'ye', 'yes'}:
            return True
        if reply in {'n', 'no', 'nop', 'nope'}:
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('invalid user response')
        print(reminder)

ask_ok("Do you want to quit?",retries=2, reminder= 'Come on, only yes or no!')
```

```python
import math, random 
from typing import List
def bubbleSort(arr: List[int]) -> List[int] :
    if len(arr) < 2:
        return
    for e in range(len(arr)):
        for i in range(e,len(arr)):
            if arr[e] > arr[i]:
                arr[e], arr[i] = arr[i], arr[e]
    
    return arr

test_arr = [random.randint(0,10) for i in range(50)]
print(test_arr)
sorted_arr_1 = sorted(test_arr)
sorted_arr_2 = bubbleSort(test_arr)
print(sorted_arr_1)
print(sorted_arr_2)
print(sorted_arr_1 == sorted_arr_2)
```
