```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.plot(x, y)
plt.grid(True)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r'$\sin(x)$', fontsize=24)
plt.xlabel(r'$x / {\rm rad}$', fontsize=24)
plt.tight_layout()
plt.savefig('SineGrid.pdf', format='pdf', dpi=1000)
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

# Define 3D figure
ax = plt.axes(projection='3d')
# vertices of the 3D polyline
zz = np.linspace(-2*np.pi, 2*np.pi, 81)

xx = np.sin(zz)
yy = np.cos(zz)

# Draw 3D polyline
ax.plot3D(xx,yy,zz,'red')

# 3D plot of scattered points
xpts = xx + 0.2*(0.5 - np.random.rand(81))
ypts = yy + 0.2*(0.5 - np.random.rand(81))
zpts = zz + 0.2*(0.5 - np.random.rand(81))
ax.scatter3D(xpts, ypts, zpts, c = ypts, cmap = 'rainbow')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Plot .coth(x) (the hyperbolic cotangent of .x) for x in the range .[−4, +4].
x = np.linspace(-4, 4, 100)
y = np.cosh(x) / np.sinh(x)
plt.plot(x, y)
plt.title("Plot of coth(x)")
plt.xlabel("x")
plt.ylabel("coth(x)")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot the function .y = 1.1 + sin(x) in the range .0 ≤ x < 4π with a logarithmic .y scale.

x = np.linspace(0, 4 * np.pi, 100)
y = 1.1 + np.sin(x)

plt.plot(x,y,'r-')
plt.yscale('log')
plt.title("Plot of y = 1.1 + sin(x) with logarithmic y scale")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Write a script that draws the two-dimensional spiral of equation (in polar coordinates) .r = 10 φ in the range .0 ≤ φ < 8π.
phi = np.linspace(0, 8 * np.pi, 1000)
r = 10 * phi
x = r * np.cos(phi)
y = r * np.sin(phi)
plt.plot(x, y)
plt.title("2D Spiral: r = 10φ")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6,6,100)
langevin = 1/np.tanh(x) - 1/x

plt.plot(x, langevin)
plt.title("Langevin Function")
plt.xlabel("x")
plt.ylabel("Langevin(x)")
plt.grid()
plt.show()
```
