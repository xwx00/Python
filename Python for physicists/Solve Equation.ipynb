```python
#!/usr/bin/env python3
from numpy import exp
import numpy as np
import matplotlib.pyplot as plt
# ................................................... Define Function
def fun(x):
    return 5.0 + 4.0 * x - exp(x)

# plot the function 
x = np.linspace(-6, 4, 400)
y = fun(x)
plt.plot(x, y, label='f(x) = 5 + 4x - exp(x)')
plt.axhline(0, color='black', lw=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')  
plt.show()


# ................... Define Interval Endpoints and Required Accuracy
a = 0.0
b = 10.0
eps = 1.0e-15

# ............................... Evaluate Function at the Endpoints
fa = fun(a)
fb = fun(b)

# ................................ Check If Bisection Method Can Work
if fa * fb > 0:
    print('wrong interval!!!', fa, fb)
    exit()

# .................................... Evaluate Solution by Bisection
iter = 1
while (b - a) > eps:
    c = (a + b) / 2.0
    fc = fun(c)
    if fc == 0:
        print('x = ', c)
        exit()
    if fc * fa > 0:
        a = c
        fa = fc
    else:
        b = c
        fb = fc
    iter += 1

# Print the final approximation
print(f'Approximate root: {c:.15f} after {iter} iterations')
```
