```python
class SingleNode:
    def __init__(self, data):
        self.data = data # store data
        self.next = None # point to the next node

class SinglyLinkedList:
    def __init__(self):
        self.head = None
    
    def append(self,data):
        new_node = SingleNode(data)
        # if the list is empty then let the new_node be the head
        if not self.head:
            self.head = new_node
            return

        # if it's not empty, we start from the head and add the new node to the end
        current = self.head
        
        while current.next:
            current = current.next
        
        current.next = new_node

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements))

    def reverse(self):
        # start from the head
        prev = None
        current = self.head
        
        while current:
            # record the next node
            next_node = current.next
            
            # reverse the linked list
            current.next = prev

            # move to the next node
            prev = current
            current = next_node

        # return head
        self.head = prev 
    
    def isPalindrome(self):
        # Edge case
        if not self.head or not self.head.next:
            return True
        
        # Standard case
        # Divide the list into two parts using slow and fast pointers
        slow, fast = self.head, self.head

        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next

        # reverse the second half list
        prev = None
        current = slow.next
        while current:
            # record the next node first
            next_node = current.next

            # let the current node point to the previous node
            current.next = prev

            # move the previous and current node to the next
            prev = current
            current = next_node
        
        first_half_pointer = self.head
        second_half_pointer = prev

        is_equal = True
        while second_half_pointer:
            if first_half_pointer.data != second_half_pointer.data:
                is_equal = False
                break
            first_half_pointer = first_half_pointer.next
            second_half_pointer = second_half_pointer.next

        # restore the linked list 
        current = prev
        prev = None
        while current:
            # record the next node first
            next_node = current.next

            # let the current node point to the previous node
            current.next = prev

            # move the previous and current node to the next
            prev = current
            current = next_node
        
        slow.next = prev

        return is_equal
    
    def partition(self, x: int):
        less_head, equal_head, greater_head = SingleNode(0),SingleNode(0),  SingleNode(0)
        less_tail, equal_tail, greater_tail = less_head, equal_head, greater_head 

        current = self.head
        while current:
            if current.data < x:
                # the original less_tail points to null or wrong node, we first update the correct next node
                less_tail.next = current
                # move the point of less_tail
                less_tail = current
            elif current.data == x:
                # the original equal_tail points to null or wrong node, we first update the correct next node
                equal_tail.next = current
                # move the point of equal_tail 
                equal_tail = current
            else:
                # the original equal_tail points to null or wrong node, we first update the correct next node
                greater_tail.next = current
                # move the point of greater_tail 
                greater_tail = current
            current = current.next
        
        # Connect three parts, connect the head and tail (æ¢å¤´æœ¯ï¼‰

        # 1. The greater_tail must point to None
        greater_tail.next = None

        # 2. The greater_head is connected the equal_tail
        # even if Greater is empty, the equal_tail will point to greater_head.next which is None type
        equal_tail.next = greater_head.next

        # 3. The less_tail points to the equal_head
        # Even if equal is empty, the equal_head.next points to the head of Greater
        # So Less is connected the Greater
        less_tail.next = equal_head.next

        # 4. Update the true head node
        self.head = less_head.next
                
        
```

```python

# --- Helper Function to Run Tests ---
def run_test(test_name, data_list, expected_result):
    # 1. Create the list
    ll = SinglyLinkedList()
    for item in data_list:
        ll.append(item)
    
    # 2. Get the actual result
    result = ll.isPalindrome()
    
    # 3. Print the report
    status = "PASS" if result == expected_result else "FAIL"
    print(f"Test: {test_name}")
    print(f"   Input: {data_list}")
    print(f"   Output: {result}")
    print(f"   Expected: {expected_result}")
    print(f"   Status: {status}")
    print("-" * 30)

# ==========================================
# TEST SUITE
# ==========================================

print("\nRunning Palindrome Tests...\n")

# 1. Basic Odd Length Palindrome
# Structure: 1 -> 2 -> 1
run_test("Odd Length Palindrome", [1, 2, 1], True)

# 2. Basic Even Length Palindrome
# Structure: 1 -> 2 -> 2 -> 1
run_test("Even Length Palindrome", [1, 2, 2, 1], True)

# 3. Non-Palindrome (Odd Length)
# Structure: 1 -> 2 -> 3
run_test("Odd Non-Palindrome", [1, 2, 3], False)

```

```python
class DoubleNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = DoubleNode(data)

        if not self.head:
            self.head = new_node
            return

        current = self.head 
        while current.next:
            current = current.next

        new_node.prev = current
        current.next = new_node
    
    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" <-> ".join(elements))
    
    def reverse(self):
        current = self.head
        prev = None

        while current:
            next_node = current.next

            current.next = prev
            current.prev = next_node

            prev = current
            current = next_node

        self.head = prev



d_list = DoublyLinkedList()
d_list.append(1)
d_list.append(2)
d_list.append(3)
d_list.append(4)
d_list.append(5)


print("Doubly Linked List:")
d_list.display()


print("Reversed Doubly Linked List:")
d_list.reverse()
d_list.display()
        
        
```
