Question: Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

Hints: Consider use yield

Solution:

```python
class Generator:
    def __init__(self,n):
        self.n = n
        self.current = 0
    
    def __iter__(self):
        next_value = self.current
        while next_value < self.n:
            yield next_value
            self.current += 7
            next_value = self.current

gen = Generator(50)
print(gen)

for i in gen:
    print(i)

print(gen.current)
```

Question A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps. The trace of robot movement is shown as the following: UP 5 DOWN 3 LEFT 3 RIGHT 2 ¡­ The numbers after the direction are steps. Please write a program to compute the distance from current position after a sequence of movement and original point. If the distance is a float, then just print the nearest integer. Example: If the following tuples are given as input to the program: UP 5 DOWN 3 LEFT 3 RIGHT 2 Then, the output of the program should be: 2

Hints: In case of input data being supplied to the question, it should be assumed to be a console input.

Solution:

```python
import math as mt

class Robot:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        self.start_x = x
        self.start_y = y

    def move(self,dx,dy):
        self.x += dx
        self.y += dy

    def distance_from_start(self):
        return round(mt.sqrt((self.x - self.start_x)**2 + (self.y - self.start_y)**2))

r = Robot(0,0)

actions = "UP 5 DOWN 3 LEFT 3 RIGHT 2".split(" ")

for i in range(0,len(actions),2):
    direction = actions[i]
    distance = int(actions[i+1])
    if direction == "UP":
        r.move(0,distance)
    elif direction == "DOWN":
        r.move(0,-distance)
    elif direction == "LEFT":
        r.move(-distance,0)
    elif direction == "RIGHT":
        r.move(distance,0)

print(f"r.x: {r.x}, r.y: {r.y}, distance from start: {r.distance_from_start()}")
```

```python
freq = {}   # frequency of words in text
line = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3." 
print(line.split())
for word in line.split():
    freq[word] = freq.get(word,0)+1

words = list(freq.keys())
words.sort()

for w in words:
    print("%s:%d" % (w,freq[w]))
```
