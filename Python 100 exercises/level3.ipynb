```python
class American(object):
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, my name is {self.name}."
    
    def printNationality(self):
        return "I am American."
    
anAmerican = American("John")
anAmerican.greet()
anAmerican.printNationality()
```

```python
class American(object):
    @staticmethod
    def printNationality():
        print("America")

anAmerican = American()
anAmerican.printNationality()
American.printNationality()
```

```python
class Circle:
    def __init__(self,r):
        self.radius = r
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
my_circle = Circle(5)
print(my_circle.area())
```

```python
class Shape(object):
    def __init__(self):
        pass
    
    def area(self):
        return 0
    
class Square(Shape):
    def __init__(self,side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
my_square = Square(4)
print(my_square.area())
print(Shape().area())
```

```python
import re
emailAddress = 'wenxuan.xie@yale.edu'
pattern = r'(?P<username>[\w.]+)@(?P<domain>\w+)\.(?P<tld>\w+)'
result = re.match(pattern, emailAddress)
print(result.group(2))
print(result.group('domain'))
print(result.groupdict())
```

```python
def fibonacci(n):
    if n < 0:
        return "Invalid input"
    elif n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci(5)
```
