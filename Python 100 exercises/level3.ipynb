```python
my_list = range(1,11)

list_filtered = filter(lambda x: x % 2 == 0, list(map(lambda x: x**2, my_list)))

print(list(list_filtered))
```

```python
class American(object):
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, my name is {self.name}."
    
    def printNationality(self):
        return "I am American."
    
anAmerican = American("John")
anAmerican.greet()
anAmerican.printNationality()
```

```python
class American(object):
    @staticmethod
    def printNationality():
        print("America")

anAmerican = American()
anAmerican.printNationality()
American.printNationality()
```

```python
class Circle:
    def __init__(self,r):
        self.radius = r
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
my_circle = Circle(5)
print(my_circle.area())
```

```python
class Shape(object):
    def __init__(self):
        pass
    
    def area(self):
        return 0
    
class Square(Shape):
    def __init__(self,side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
my_square = Square(4)
print(my_square.area())
print(Shape().area())
```

```python
def throws():
    return 5/0

try:
    throws()
except ZeroDivisionError:
    print("division by zero!")
except Exception:
    print('Caught an exception')
finally:
    print('In finally block for cleanup')
```

```python
import re

pattern = '(?P<username>\w+)@(?P<companyname>\w+)\.(?P<domain>\w+)'
search_result = re.match(pattern, 'john@google.com')

print(search_result.groupdict())
print(search_result['username'])
```
